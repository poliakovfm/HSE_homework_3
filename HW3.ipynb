```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Зададим масивы с названиями и со значениями
x = ["a","b","c","d","e","f","g"]
y1 = np.array([3,8,6,4,8,5,2])
y2 = np.array([10,2,4,3,5,6,15])
y3 = np.array([5,6,2,5,5,2,8])
y4 = np.array([1,2,3,4,5,6,7])

# Нормалиризуем наши значения
snum = y1+y2+y3+y4
y1 = y1/snum*100.
y2 = y2/snum*100.
y3 = y3/snum*100.
y4 = y4/snum*100.

# Зададим высоту и ширину графика
plt.figure(figsize=(4,7))

# Зададим наши bar'ы 
plt.bar(x, y1, label='y1')
plt.bar(x, y2, bottom=y1,label='y2')
plt.bar(x, y3, bottom=y1+y2,label='y3')
plt.bar(x, y4, bottom=y1+y2+y3,label='y4')

# Добавим текст, соответствующий перцентилю данных в bar'е
for xpos, ypos, yval in zip(x, y1/2, y1):
    plt.text(xpos, ypos, "%.1f"%yval, ha="center", va="center")
for xpos, ypos, yval in zip(x, y1+y2/2, y2):
    plt.text(xpos, ypos, "%.1f"%yval, ha="center", va="center")
for xpos, ypos, yval in zip(x, y1+y2+y3/2, y3):
    plt.text(xpos, ypos, "%.1f"%yval, ha="center", va="center")
for xpos, ypos, yval in zip(x, y1+y2+y3+y4/2, y4):
    plt.text(xpos, ypos, "%.1f"%yval, ha="center", va="center")


# Присвоим названия оси Ox и всему графику в целом
plt.xlabel("Group")
plt.title('Some random data')

# Выведем график
plt.show()
```
